# 2. 결측치 처리하기

## 1) 필요없는 컬럼 삭제하기

*   OFF_STREET_NAME
*   DRIVER_LICENSE_JURISDICTION
*   VEHICLE_MAKE
*   항목 추가





```python
#data.drop(columns=['OFF_STREET_NAME'], inplace=True)
data.drop(columns=['DRIVER_LICENSE_JURISDICTION'], inplace=True)
data.drop(columns=['VEHICLE_MAKE'], inplace=True)
```

## 2)  POINT_OF_IMPACT > VEHICLE_DAMAGE > VEHICLE_DAMAGE_1

POINT_OF_IMPACT 컬럼의 값을 VEHICLE_DAMAGE 컬럼의 값을 VEHICLE_DAMAGE_1 컬럼의 결측값에 채우기

```python
data['VEHICLE_DAMAGE'] = data['VEHICLE_DAMAGE'].fillna(data['POINT_OF_IMPACT'])

# 결과 확인
print(data[['POINT_OF_IMPACT', 'VEHICLE_DAMAGE']].head())
```

```python
data['VEHICLE_DAMAGE_1'] = data['VEHICLE_DAMAGE_1'].fillna(data['VEHICLE_DAMAGE'])

# 결과 확인
print(data[['VEHICLE_DAMAGE', 'VEHICLE_DAMAGE_1']].head())
```

```python
# 모든 컬럼의 데이터값의 고유값 과 각각의 개수 출력

# 사용할 컬럼 리스트
columns = ['POINT_OF_IMPACT','VEHICLE_DAMAGE', 'VEHICLE_DAMAGE_1']

# 각 컬럼에 대해 고유값과 개수 출력
for column in columns:
    if column in data.columns:
        value_counts = data[column].value_counts().sort_index()
        sorted_unique_values = sorted(value_counts.items())

        print(f"컬럼명: {column}")
        print(f"정렬된 unique 값: {[f'{value} : {count}' for value, count in sorted_unique_values]}\n")
    else:
        print(f"컬럼명: {column}는 데이터프레임에 존재하지 않습니다.\n")
```

## 3)  DRIVER_LICENSE_STATUS 컬럼의 NaN 값을 'Licensed'로 채우기

채우기 전 -
컬럼명: DRIVER_LICENSE_STATUS
정렬된 unique 값:
<br> ['Licensed : 2005230', 'Permit : 18170', 'Unlicensed : 40808']


```python
# DRIVER_LICENSE_STATUS 컬럼의 NaN 값을 'Licensed'로 채우기
data['DRIVER_LICENSE_STATUS'] = data['DRIVER_LICENSE_STATUS'].fillna('Licensed')

# unique 값과 개수 출력
unique_values = data['DRIVER_LICENSE_STATUS'].value_counts()
print("Unique 값과 개수:")
print(unique_values)
```

## 4) VEHICLE_OCCUPANTS 이상치 제거 및 평균값으로 채우기

```python
# 100 이상의 값을 가지는 행 필터링
abnormal_values = data[data['VEHICLE_OCCUPANTS'] >= 80]

# VEHICLE_OCCUPANTS를 기준으로 낮은 값이 위로 오도록 정렬하여 출력
sorted_abnormal_values = abnormal_values.sort_values(by='VEHICLE_OCCUPANTS', ascending=True)

# DataFrame 형식으로 출력
from IPython.display import display
display(sorted_abnormal_values)
```

## 5) POINT_OF_IMPACT, TRAVEL_DIRECTION, VEHICLE_TYPE_CODE_2, VEHICLE_TYPE_CODE_1  unknown으로 채우기

```python
columns_to_fill = ['POINT_OF_IMPACT', 'TRAVEL_DIRECTION', 'VEHICLE_TYPE_CODE_2', 'VEHICLE_TYPE_CODE_1', 'CONTRIBUTING_FACTOR_1', 'CONTRIBUTING_FACTOR_2']

# NaN 값을 'unknown'으로 채우기
data[columns_to_fill] = data[columns_to_fill].fillna('unknown')

# 결과 확인
print(data[columns_to_fill].head())
```

## 6) CONTRIBUTING_FACTOR_VEHICLE_2, CONTRIBUTING_FACTOR_VEHICLE_1  unspecified로 채우기

```python
columns_to_fill = ['CONTRIBUTING_FACTOR_VEHICLE_2', 'CONTRIBUTING_FACTOR_VEHICLE_1']

# NaN 값을 'unknown'으로 채우기
data[columns_to_fill] = data[columns_to_fill].fillna('unspecified')

# 결과 확인
print(data[columns_to_fill].head())
```

## 7) VEHICLE_TYPE_CODE_1 로 VEHICLE_TYPE 채우기

```python
data['VEHICLE_TYPE'] = data['VEHICLE_TYPE'].fillna(data['VEHICLE_TYPE_CODE_1'])

# 결과 확인
print(data[['VEHICLE_TYPE_CODE_1', 'VEHICLE_TYPE']].head())
```

## 8) 'BOROUGH', 'LOCATION' 둘 다 비어있는 행 삭제하기

```python
# 'BOROUGH' 및 'LOCATION' 컬럼이 모두 NaN 값을 가지는 행 삭제
data.dropna(subset=['BOROUGH', 'LOCATION'], how='all', inplace=True)

# 결과 확인
print(data[['BOROUGH', 'LOCATION']].head())
```
